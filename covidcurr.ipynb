```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
plt.ion
```

```python
data=pd.read_csv(r"worldometer_snapshots_April18_to_May18.csv")
df=pd.DataFrame(data)
df
```

```python
df.set_index("Date")
df
```

```python
coun=input("Country Name\n")
```

```python
newd=df.loc[(df.Country == coun)]
newd
```

```python
X=newd["Date"]
X
```

```python
Y=newd["Total Cases"]
Y
```

```python
plt.title("TOTAL CASES ON EACH DAY")
plt.ylabel("No. of Cases")
plt.xlabel("Date")
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.bar(X,Y,color="#2E4053")
plt.show()
```

```python
A=newd["Total Deaths"]
A
```

```python
plt.figure(figsize=(10,10))
plt.title("TOTAL DEATHS Due to Corona Virus ON EACH DAY")
plt.ylabel("No. of Deaths",fontweight="bold")
plt.xlabel("Date",fontweight="bold")
plt.xticks(rotation=85)
plt.grid(axis="y")
plt.stem(X,A)
plt.show()
```

```python
B=newd["Total Recovered"]
B
```

```python
plt.title("PERSONS RECOVERED ON EACH DAY")
plt.ylabel("No. of Persons",fontweight="bold")
plt.xlabel("Date",fontweight="bold")
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.plot(X,B,color="purple")
plt.show()
```

```python
C=newd["Active Cases"]
C
```

```python
plt.title("ACTIVE CASES ON EACH DAY")
plt.ylabel("No. of Cases",fontweight="bold")
plt.xlabel("Date",fontweight="bold")
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.bar(X,C)
plt.show()
```

```python
D=newd["Total Tests"]
D
```

```python
plt.title("TOTAL TESTS DONE ON EACH DAY")
plt.ylabel("No. of TESTS",fontweight="bold")
plt.xlabel("Date",fontweight="bold")
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.stem(X,D,"dk")
plt.show()
```
